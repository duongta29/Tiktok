```python
from PIL import Image

def rotate_image(image_path, angle):
    # Load h√¨nh ·∫£nh
    image = Image.open(image_path)

    # Chuy·ªÉn ƒë·ªïi ch·∫ø ƒë·ªô m√†u t·ª´ RGBA sang RGB
    image = image.convert("RGB")

    # Xoay ·∫£nh
    rotated_image = image.rotate(angle)

    # L∆∞u ·∫£nh ƒë√£ xoay
    rotated_image.save("rotated_image.jpg")


# G·ªçi h√†m xoay ·∫£nh v·ªõi g√≥c ƒë·ªô c·ª• th·ªÉ (v√≠ d·ª•: 45 ƒë·ªô)
rotate_image("rotateCaptcha\outer3.jpg", 65)
```

```python
import cv2
import numpy as np

# ƒê∆∞·ªùng d·∫´n ƒë·∫øn ·∫£nh
image_path = "rotateCaptcha\inner11.jpg"

# ƒê·ªçc ·∫£nh v√†o d∆∞·ªõi d·∫°ng grayscale
image = cv2.imread(image_path, 0)

# T√¨m vi·ªán ƒëi·ªÉm v√† b√°n k√≠nh c·ªßa h√¨nh tr√≤n
contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
center, radius = cv2.minEnclosingCircle(contours[0])
center = tuple(map(int, center))
radius = int(radius)

# T·∫°o ·∫£nh tr·∫Øng c√≥ k√≠ch th∆∞·ªõc gi·ªëng v·ªõi ·∫£nh g·ªëc
mask = np.zeros_like(image)

# V·∫Ω h√¨nh tr√≤n l√™n ·∫£nh tr·∫Øng
cv2.circle(mask, center, radius, (255, 255, 255), -1)

# C·∫Øt ·∫£nh g·ªëc th√†nh hai ph·∫ßn: h√¨nh tr√≤n nh·ªè v√† ph·∫ßn c√≤n l·∫°i
small_circle = cv2.bitwise_and(image, image, mask=mask)
outer_arc = cv2.bitwise_and(image, image, mask=cv2.bitwise_not(mask))

# Hi·ªÉn th·ªã ·∫£nh g·ªëc, h√¨nh tr√≤n nh·ªè v√† ph·∫ßn c√≤n l·∫°i
cv2.imshow("Original Image", image)
cv2.imshow("Small Circle", small_circle)
cv2.imshow("Outer Arc", outer_arc)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

def rotate_image(image_path, angle):
    # ƒê·ªçc ·∫£nh v√† chuy·ªÉn ƒë·ªïi sang ·∫£nh x√°m
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Kh·ªüi t·∫°o ƒë·ªëi t∆∞·ª£ng SIFT
    sift = cv2.SIFT_create()

    # T√¨m c√°c ƒëi·ªÉm ƒë·∫∑c tr∆∞ng tr√™n ·∫£nh g·ªëc
    keypoints, descriptors = sift.detectAndCompute(gray, None)

    # T·∫°o ma tr·∫≠n xoay
    center = (image.shape[1] // 2, image.shape[0] // 2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

    # Xoay ·∫£nh
    rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))

    return rotated_image

# G·ªçi h√†m xoay ·∫£nh v·ªõi g√≥c ƒë·ªô c·ª• th·ªÉ (v√≠ d·ª•: 45 ƒë·ªô)
rotated_image = rotate_image("rotateCaptcha\inner2.jpg", 25)

# Hi·ªÉn th·ªã ·∫£nh g·ªëc v√† ·∫£nh ƒë√£ xoay
cv2.imshow("Original Image", cv2.imread("rotateCaptcha\inner2.jpg"))
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
